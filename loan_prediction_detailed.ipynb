```python
data = pd.read_csv('/content/loan_dataset.csv')
```

```python
data = data.drop('loan_id',axis=1)
```

```python
data.rename(columns = {' no_of_dependents':'no_of_dependents',' education':'education',' self_employed':'self_employed',' income_annum':'income_annum',' loan_amount':'loan_amount',' loan_term':'loan_term',' cibil_score':'cibil_score',' residential_assets_value':'residential_assets_value',' commercial_assets_value':'commercial_assets_value',' luxury_assets_value':'luxury_assets_value',' bank_asset_value':'bank_asset_value',' loan_status':'loan_status'},inplace=True)
```

```python
data['no_of_dependents'].unique()
```

```python
data.head(5)
```

```python
data['education'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()

data["education"] = LE.fit_transform(data["education"])
data["self_employed"] = LE.fit_transform(data["self_employed"])
data["loan_status"] = LE.fit_transform(data["loan_status"])
```

```python
data.head(5)
```

```python
x = data.drop('loan_status',axis=1)
y=data['loan_status']
```

```python
data.describe().style.background_gradient(cmap="Blues")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


import seaborn as sns
sns.set_style('dark')
```

```python
plt.figure(figsize=(10, 5))
plt.grid()
sns.scatterplot(x=data['cibil_score'],y=data['loan_status'],data=data)
plt.title('Relation between CIBIL score and loan status')
plt.xlabel('Cibil Score')
plt.ylabel('Loan status')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.figure(figsize=(10, 5))
plt.plot(data['income_annum'], data['loan_amount'])

plt.title('Relation between Annual Income and Loan Amount')
plt.xlabel('Income')
plt.ylabel('Loan Amount')

formatter = ticker.ScalarFormatter(useOffset=False)
formatter.set_scientific(False)

plt.gca().xaxis.set_major_formatter(formatter)
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## **LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x_train,y_train)
```

```python
log_pred = log_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,accuracy_score
print('Accuracy of Logistic regression model:',accuracy_score(log_pred,y_test)*100)
print('Mean absolute error:',(mean_absolute_error(log_pred,y_test)))
print('Root mean squared error:',np.sqrt(mean_squared_error(log_pred,y_test)))
print('R2_score:',r2_score(log_pred,y_test))
```

## **SVC**

```python
from sklearn import svm
svm_model = svm.SVC()
svm_model.fit(x_train,y_train)
```

```python
svm_pred = svm_model.predict(x_test)
```

```python
print('Accuracy of SVM model:',accuracy_score(svm_pred,y_test)*100)
print('Mean absolute error:',(mean_absolute_error(svm_pred,y_test)))
print('Root mean squared error:',np.sqrt(mean_squared_error(svm_pred,y_test)))
print('R2_score:',r2_score(svm_pred,y_test))
```

## **Descision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
```

```python
dt_pred = dt_model.predict(x_test)
```

```python
print('Accuracy of Decision Tree Classifier model:',accuracy_score(dt_pred,y_test)*100)
print('Mean absolute error:',(mean_absolute_error(dt_pred,y_test)))
print('Root mean squared error:',np.sqrt(mean_squared_error(dt_pred,y_test)))
print('R2_score:',r2_score(dt_pred,y_test))
```

## **Rendom forest classifier**

```python
from sklearn.ensemble import RandomForestClassifier
rf_model =RandomForestClassifier()
rf_model.fit(x_train,y_train)
```

```python
rf_pred = rf_model.predict(x_test)
```

```python
print('Accuracy of Random forest Classifier model:',accuracy_score(rf_pred,y_test)*100)
print('Mean absolute error:',(mean_absolute_error(rf_pred,y_test)))
print('Root mean squared error:',np.sqrt(mean_squared_error(rf_pred,y_test)))
print('R2_score:',r2_score(rf_pred,y_test))
```

## **Gradient Boosting Classifier**

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_model =GradientBoostingClassifier()
gb_model.fit(x_train,y_train)
```

```python
gb_pred = gb_model.predict(x_test)
```

```python
print('Accuracy of Gradient Boosting Classifier model:',accuracy_score(gb_pred,y_test)*100)
print('Mean absolute error:',(mean_absolute_error(gb_pred,y_test)))
print('Root mean squared error:',np.sqrt(mean_squared_error(gb_pred,y_test)))
print('R2_score:',r2_score(gb_pred,y_test))
```
